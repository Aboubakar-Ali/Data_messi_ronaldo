## Comparaison du nombre de saisons jouées par Messi et Ronaldo

Ici, nous retrouvons un graphique representatif des saisons ayant chaque joueur

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données depuis le fichier CSV dans un DataFrame
data = pd.read_csv("../datsets/data.csv")

# Compter le nombre de saisons de Messi et Ronaldo
saisons_messi = len(data[data["Player"] == "Lionel Messi"]["Season"].unique())
saisons_ronaldo = len(data[data["Player"] == "Cristiano Ronaldo"]["Season"].unique())

print("Messi a joué dans {} saisons.".format(saisons_messi))
print("Ronaldo a joué dans {} saisons.".format(saisons_ronaldo))

# Créer un graphique en barres pour comparer le nombre de saisons jouées
joueurs = ["Lionel Messi", "Cristiano Ronaldo"]
saisons = [saisons_messi, saisons_ronaldo]

plt.bar(joueurs, saisons)
plt.xlabel("Joueur")
plt.ylabel("Nombre de saisons")
plt.title("Comparaison du nombre de saisons jouées par Messi et Ronaldo")
plt.show()
```
