```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données depuis le fichier CSV dans un DataFrame
data = pd.read_csv("../datsets/data.csv")

# Convertir la colonne "Date" en objet datetime
data["Date"] = pd.to_datetime(data["Date"], format='%Y-%m-%d')

# Transformer les types de buts en chiffres
goals_dict = {'Solo run': 1, 'Header': 1, 'Right-footed shot': 1, 'Left-footed shot': 1, 'Direct free kick': 1, 
              'Tap-in': 1, 'Penalty': 1, 'Penalty rebound': 1, 'Deflected shot on goal': 1, 'Counter attack goal': 1, 
              'Long distance kick': 1, 'Chest': 1}
data["Goals"] = data["Type"].replace(goals_dict)

# Regrouper les données par joueur et par saison, et sommer le nombre de buts
goals_by_player_and_season = data.groupby(["Player", pd.Grouper(key="Date", freq="A")])["Goals"].sum()

# Créer un graphique avec deux courbes pour Messi et Ronaldo
goals_by_player_and_season.loc["Lionel Messi"].plot(label="Lionel Messi")
goals_by_player_and_season.loc["Cristiano Ronaldo"].plot(label="Cristiano Ronaldo")
plt.legend()
plt.title("Nombre de buts marqués par saison et par joueur")
plt.xlabel("Saison")
plt.ylabel("Nombre de buts")
plt.show()


```
