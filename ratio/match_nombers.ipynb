## Nombre de matchs jou√©s par joueur

Ici, nous retrouvons un graphique representatif des buts de chauque joueur!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donn√©es depuis le fichier CSV dans un DataFrame
data = pd.read_csv("./datsets/data.csv")

# Calculer le nombre de matchs jou√©s pour Messi
nb_matchs_messi = len(data[data["Player"] == "Lionel Messi"])

# Calculer le nombre de matchs jou√©s pour Ronaldo
nb_matchs_ronaldo = len(data[data["Player"] == "Cristiano Ronaldo"])

# Cr√©er un graphique en barres pour comparer les nombres de matchs jou√©s
joueurs = ["Lionel Messi", "Cristiano Ronaldo"]
nb_matchs = [nb_matchs_messi, nb_matchs_ronaldo]

print("nb_messi = ",nb_matchs_messi)
print("nb_ronaldo = ",nb_matchs_ronaldo)
plt.bar(joueurs, nb_matchs)
plt.title("Nombre de matchs jou√©s par joueur")
plt.xlabel("Joueur")
plt.ylabel("Nombre de matchs")
plt.show()
```
