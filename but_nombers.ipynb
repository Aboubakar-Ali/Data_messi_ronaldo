```python
import pandas as pd
import matplotlib.pyplot as plt


# Charger les données depuis le fichier CSV dans un DataFrame
data = pd.read_csv("data.csv")

# Compter le nombre de buts pour chaque joueur
but_messi = data[data["Player"] == "Lionel Messi"]["Type"].value_counts().sum()
but_ronaldo = data[data["Player"] == "Cristiano Ronaldo"]["Type"].value_counts().sum()

print("Messi a marqué {} buts.".format(but_messi))
print("Ronaldo a marqué {} buts.".format(but_ronaldo))

# Créer un graphique à barres
plt.bar(["Messi", "Ronaldo"], [but_messi, but_ronaldo], color=["#fcba03", "#7f2626"])

# Ajouter un titre et des étiquettes d'axe
plt.title("Nombre de buts de Messi et Ronaldo")
plt.xlabel("Joueur")
plt.ylabel("Nombre de buts")

# Afficher le graphique
plt.show()
```
