## Comparaison des récompenses entre Messi et Ronaldo

Le graphique cumulé représente le nombre de victoires de Cristiano Ronaldo et Lionel Messi pour trois titres différents : le Ballon d'Or, le Soulier d'Or et le Best FIFA Men's Player. Chaque barre représente une récompense spécifique, et la hauteur de la barre représente le nombre de victoires pour chaque joueur. Le graphique permet une comparaison rapide et facile entre les deux joueurs pour chaque titre. Les couleurs utilisées sont le rouge et le jaune, respectivement associés à Ronaldo et Messi, pour une identification visuelle rapide.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données depuis le fichier CSV dans un DataFrame
data = pd.read_csv("../datsets/ballon_dor.csv")
data1 = pd.read_csv("../datsets/soulier_dor.csv")
data2 = pd.read_csv("../datsets/the_best.csv")


# Filtrer les données pour ne contenir que Ronaldo et Messi
ronaldo_data = data[data['Nom'] == 'Cristiano Ronaldo']
messi_data = data[data['Nom'] == 'Lionel Messi']
ronaldo_data1 = data1[data1['Nom'] == 'Cristiano Ronaldo']
messi_data1 = data1[data1['Nom'] == 'Lionel Messi']
ronaldo_data2 = data2[data2['Nom'] == 'Cristiano Ronaldo']
messi_data2 = data2[data2['Nom'] == 'Lionel Messi']


#recupération des données
bo_ronaldo = ronaldo_data[ronaldo_data["Nom"] == "Cristiano Ronaldo"].value_counts().sum()
bo_messi = messi_data[messi_data["Nom"] == "Lionel Messi"].value_counts().sum()
so_ronaldo = ronaldo_data1[ronaldo_data1["Nom"] == "Cristiano Ronaldo"].value_counts().sum()
so_messi = messi_data1[messi_data1["Nom"] == "Lionel Messi"].value_counts().sum()
best_ronaldo = ronaldo_data2[ronaldo_data2["Nom"] == "Cristiano Ronaldo"].value_counts().sum()
best_messi = messi_data2[messi_data2["Nom"] == "Lionel Messi"].value_counts().sum()



# Création du graphique avec Seaborn
sns.set_style("darkgrid")
plt.figure(figsize=(8,6))
sns.barplot(x=["Ballon d'Or Ronaldo", "Ballon d'Or Messi", "Soulier d'Or Ronaldo", "Soulier d'Or Messi", "The Best Ronaldo", "The Best Messi"], 
            y=[bo_ronaldo, bo_messi, so_ronaldo, so_messi, best_ronaldo, best_messi], 
            palette=["#960018", "#fcbb1f"])
plt.title("Comparaison des récompenses entre Messi et Ronaldo")
plt.xlabel("Récompenses")
plt.ylabel("Nombre de victoires")
plt.xticks(rotation=45, ha="right")
plt.show()
```

## Pourcentage des titres remportés par Cristiano Ronaldo et Lionel Messi


Le cercle ci-dessous représente la répartition des titres remportés par les joueurs de football Cristiano Ronaldo et Lionel Messi en club. Chaque section du cercle représente le pourcentage de titres remportés par chaque joueur.

```python
import pandas as pd
import matplotlib.pyplot as plt

data3 = pd.read_csv("../datsets/joeurs_titrés.csv")

# Filtrer les données pour ne contenir que Ronaldo et Messi
ronaldo_data3 = data3[data3['Nom'] == 'Cristiano Ronaldo']
messi_data3 = data3[data3['Nom'] == 'Lionel Messi']

# nombre de titre que possede chacun
titre_ronaldo = ronaldo_data3["Nombre de Trophée"].sum()
titre_messi = messi_data3["Nombre de Trophée"].sum()

# Somme des titres de ronaldo et celui de messi
total_titres = titre_ronaldo + titre_messi

# calcul de pourcentage de titre remporté par rapport à la Somme des titres des deux 
pourcentage_ronaldo = round((titre_ronaldo / total_titres) * 100, 2)
pourcentage_messi = round((titre_messi / total_titres) * 100, 2)

# Création d'un diagramme circulaire
labels = ['Cristiano Ronaldo', 'Lionel Messi']
sizes = [pourcentage_ronaldo, pourcentage_messi]
colors = ['red', 'blue']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%', startangle=90)

# Dessiner un cercle blanc au centre du diagramme pour créer un trou
centre_cercle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_cercle)

# Ajouter une légende et un titre
ax1.axis('equal')
plt.legend(labels)
plt.title('Pourcentage des titres remportés par Cristiano Ronaldo et Lionel Messi')

# Afficher le diagramme
plt.show()
```
