## RÃ©partition des buts de Messi par club

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donnÃ©es depuis le fichier CSV dans un DataFrame
data = pd.read_csv("../datsets/data.csv")

# Transformer les types de buts en chiffres
goals_dict = {'Solo run': 1, 'Header': 1, 'Right-footed shot': 1, 'Left-footed shot': 1, 'Direct free kick': 1, 
              'Tap-in': 1, 'Penalty': 1, 'Penalty rebound': 1, 'Deflected shot on goal': 1, 'Counter attack goal': 1, 
              'Long distance kick': 1, 'Chest': 1}
data["Goals"] = data["Type"].replace(goals_dict)

# Filtrer les donnÃ©es pour ne garder que les buts de Messi
ronaldo_data = data[data["Player"] == "Lionel Messi"]

# Regrouper les donnÃ©es par club et sommer le nombre de buts
goals_by_club = ronaldo_data.groupby("Club")["Goals"].sum().sort_values(ascending=False)

# CrÃ©er un graphique en forme de cercle
plt.pie(goals_by_club.values, labels=goals_by_club.index, autopct='%1.1f%%')
plt.title("RÃ©partition des buts de Messi par club")
plt.show()
```
