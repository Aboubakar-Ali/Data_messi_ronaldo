## RÃ©partition des buts de Ronaldo et Messi par mi-temps

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donnÃ©es depuis le fichier CSV dans un DataFrame
data = pd.read_csv("../datsets/data.csv")

# Filtrer les donnÃ©es pour ne contenir que Ronaldo et Messi
ronaldo_data = data[data['Player'] == 'Cristiano Ronaldo']
messi_data = data[data['Player'] == 'Lionel Messi']

# Convertir les minutes de buts en entiers
ronaldo_data['Minute'] = ronaldo_data['Minute'].apply(lambda x: int(x.split('+')[0]) + int(x.split('+')[1]) if '+' in x else int(x))
messi_data['Minute'] = messi_data['Minute'].apply(lambda x: int(x.split('+')[0]) + int(x.split('+')[1]) if '+' in x else int(x))

# SÃ©parer les minutes de buts de Ronaldo en premiÃ¨re et deuxiÃ¨me mi-temps
ronaldo_first_half = ronaldo_data[ronaldo_data['Minute'] < 46]
ronaldo_second_half = ronaldo_data[ronaldo_data['Minute'] >= 46]

# SÃ©parer les minutes de buts de Messi en premiÃ¨re et deuxiÃ¨me mi-temps
messi_first_half = messi_data[messi_data['Minute'] <= 45]
messi_second_half = messi_data[messi_data['Minute'] > 45]

# CrÃ©er un histogramme pour les minutes de buts de Ronaldo et Messi
plt.hist([ronaldo_first_half['Minute'], ronaldo_second_half['Minute'], messi_first_half['Minute'], messi_second_half['Minute']], 
         color=['red', 'red', 'blue', 'blue'],
         label=['Ronaldo', '', 'Messi', ''])

# Ajouter une lÃ©gende et des Ã©tiquettes d'axe
plt.legend()
plt.xlabel('Minute de but')
plt.ylabel('Nombre de buts')
plt.title('RÃ©partition des buts de Ronaldo et Messi par mi-temps')

# Afficher le graphique
plt.show()
```
