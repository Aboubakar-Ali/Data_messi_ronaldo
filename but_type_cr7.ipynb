## RÃ©partition des buts de Ronaldo par type

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donnÃ©es depuis le fichier CSV dans un DataFrame
data = pd.read_csv("./datsets/data.csv")

# Filtrer les donnÃ©es pour ne garder que les buts de Ronaldo
ronaldo_data = data[data["Player"] == "Cristiano Ronaldo"]

# Compter le nombre de buts par type de but
goals_by_type = ronaldo_data["Type"].value_counts()

# CrÃ©er le graphique en forme de cercle avec les pourcentages
plt.pie(goals_by_type.values, labels=goals_by_type.index, autopct='%1.1f%%')
plt.title("RÃ©partition des buts de Ronaldo par type")
plt.show()

```
