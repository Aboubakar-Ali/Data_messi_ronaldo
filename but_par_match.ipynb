## Ratio de buts par match

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données depuis le fichier CSV dans un DataFrame
data = pd.read_csv("./datsets/data.csv")

# Calculer le nombre de buts et de matchs pour chaque joueur
messi_stats = data[data["Player"] == "Lionel Messi"]
ronaldo_stats = data[data["Player"] == "Cristiano Ronaldo"]
but_messi = messi_stats["Type"].value_counts().sum()
but_ronaldo = ronaldo_stats["Type"].value_counts().sum()
matchs_messi = messi_stats["Player"].count()
matchs_ronaldo = ronaldo_stats["Player"].count()

# Calculer le ratio de buts par match pour chaque joueur
ratio_messi = but_messi / matchs_messi
ratio_ronaldo = but_ronaldo / matchs_ronaldo

# Print ration
print("messi: " + str(ratio_messi) + "/match")
print("Ronaldo: " + str(ratio_ronaldo) + "/match")
# Créer le graphique
plt.bar(["Messi", "Ronaldo"], [ratio_messi, ratio_ronaldo], color='pink')
plt.title("Ratio de buts par match")
plt.xlabel("Joueur")
plt.ylabel("Ratio")
plt.ylim(0, 1.5)
plt.show()
```
